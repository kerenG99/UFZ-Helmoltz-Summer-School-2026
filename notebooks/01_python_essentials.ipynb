## 2. Python as a calculator — arithmetic operators
| op | meaning | | op | meaning |
|----|---------|-|----|---------|
| `+` `-` `*` | add, subtract, multiply | | `/` | float division |
| `//` | floor (integer) division | | `%` | modulo (remainder) |
| `**` | power | | | |

````python
# The everyday operators behave as you expect:
print(7 + 3, 7 - 3, 7 * 3)   # addition, subtraction, multiplication

# Two kinds of division - this trips people up:
print(7 / 2)      # '/'  -> 3.5  ALWAYS gives a float (decimal)
print(7 // 2)     # '//' -> 3    floor division: drops the remainder
print(7 % 3)      # '%'  -> 1    modulo: the remainder itself

# Why care about '%'? Reading frames: position % 3 tells you the codon slot.
print(2 ** 10)    # '**' -> 1024  power (2 to the 10th)
````

Operators follow the usual **precedence** (`**` before `*`/`/` before `+`/`-`); use parentheses when in doubt.

````python
# Same numbers, different grouping -> different result. Parentheses win.
print(2 + 3 * 4)      # multiplication first  -> 14
print((2 + 3) * 4)    # parentheses first      -> 20
````

## 3. Variables & values

## 4. Data types
Every value has a **type**. `type()` reveals it.

| type | example | biological use |
|------|---------|----------------|
| `int` | `1465` | a length, a read count |
| `float` | `0.53` | GC content, pH, abundance |
| `str` | `'ACGT'` | a DNA/RNA/protein sequence |
| `bool` | `True`/`False` | result of a test |

````python
# type() tells you what kind of value you are holding.
# Getting types wrong is the #1 source of beginner bugs, so check often.
print(type(sequence_len))   # <class 'int'>
print(type(gc_fraction))    # <class 'float'>
print(type(gene_id))        # <class 'str'>
print(type(is_archaeon))    # <class 'bool'>
````

### Type conversion
Data read from files often arrives as **text**; convert it explicitly before doing maths on it.

````python
# A number read from a CSV file arrives as TEXT, not a number:
count_text = '128'          # note the quotes -> this is a string

# int() converts text -> integer so we can do arithmetic:
count = int(count_text)
print(count + 2, type(count))   # 130 <class 'int'>

# The other direction and float conversion:
print(float('0.53'))        # text -> float  (0.53)
print(str(1465))            # number -> text ('1465')
````

## 5. Relational & logical operators
Comparisons return a **bool**. Python even allows *chained* comparisons.

````python
# Relational operators ASK a question; the answer is True or False.
temp = 22
print(temp == 22)          # '==' equal to?      (note: two equals signs!)
print(temp != 15)          # '!=' not equal to?
print(temp > 25)           # greater than?
print(15 <= temp <= 25)    # chained: is temp between 15 and 25 inclusive?
````

````python
# Logical operators COMBINE True/False values.
high_gc = gc_fraction > 0.5     # a bool we computed above
is_long = sequence_len > 1000
print(high_gc and is_long)      # 'and' -> True only if BOTH are true
print(high_gc or is_archaeon)   # 'or'  -> True if AT LEAST ONE is true
print(not is_archaeon)          # 'not' -> flips True<->False
````

## 6. Strings hold sequences
A DNA sequence is just text. Strings support **membership** (`in`), **concatenation** (`+`) and **repetition** (`*`).

````python
# Store a real-ish 16S fragment. In Python, a DNA sequence is a STRING.
seq = 'AGAGTTTGATCCTGGCTCAGATTGAACGCTGGCGGCAGGCCTAACACATGCAAGTCGAACGG'

print('length          : ', len(seq))         # len() = number of bases
print('has ATG start?  : ', 'ATG' in seq)      # 'in' = membership test
print('GGGG absent     : ', 'GGGG'  in seq) # 'not in' = the opposite
print('concatenation   : ', 'ATG' + seq[:6])   # '+' glues strings together
print('primer x3       : ', 'AT' * 3)          # '*' repeats a string
````

## 7. Base composition & GC content
Count each base and compute GC content with only operators and `.count()`:

$$GC = \frac{G + C}{\text{length}}$$

````python
# .count('X') counts how many times 'X' appears in the string.
# We unpack the four counts into four variables in one line:
nA, nC, nG, nT = seq.count('A'), seq.count('C'), seq.count('G'), seq.count('T')
print(f'A={nA}  C={nC}  G={nG}  T={nT}')

# GC content = (G + C) / total length. That's the whole formula.
gc = (nG + nC) / len(seq)
print(f'GC content = {gc:.1%}')   # the ':.1%' formats a fraction as a percent
````
